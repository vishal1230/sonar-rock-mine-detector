```python
!pip install Gradio
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import gradio as gr

# Load and train model
data = pd.read_csv('sonar.csv', header=None)
X = data.drop(columns=60, axis=1)
Y = data[60]
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, stratify=Y, random_state=1)
model = LogisticRegression()
model.fit(X_train, Y_train)

# Prediction function
def predict_mine_rock(*inputs):
    input_np = np.array(inputs).reshape(1, -1)
    prediction = model.predict(input_np)[0]
    return "Rock" if prediction == 'R' else "Mine"

# Gradio UI
inputs = [gr.Number(label=f"Feature {i+1}") for i in range(60)]
app = gr.Interface(fn=predict_mine_rock, inputs=inputs, outputs="text")
app.launch()
```
